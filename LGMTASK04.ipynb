IMPORTING LIBRARIES

```python
import cv2
from google.colab.patches import cv2_imshow
```

READ THE ORIGINAL IMAGE

```python
image = cv2.imread("ganuuuu.jpeg")
```

```python
cv2_imshow(image)
```

CONVERT RBG IMAGE TO GREY SCALE

```python
grey_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(grey_img)
```

INVERT THE GREYSCALE IMAGE

```python
invert = cv2.bitwise_not(grey_img)
```

```python
cv2_imshow(invert)
```

APPLYING GAUSSIAN BLUR

```python
blur = cv2.GaussianBlur(invert,(21,21),0)
```

```python
cv2_imshow(blur)
```

BLEND THE IMAGE WITH BLUR NEGATIVE



```python
invertedblur = cv2.bitwise_not(blur)
```

```python
cv2_imshow(invertedblur)
```

```python
sketch = cv2.divide(grey_img,invertedblur,scale=146.0)
```

OUTPUT IMAGE

```python
cv2_imshow(sketch)
```
